```python
import matplotlib.pyplot as plt
import seaborn as sns

```

```python

output_dir = 'visualizations/'
os.makedirs(output_dir, exist_ok=True)

sns.set(style="whitegrid")

```

```python
# 1. Histogram of Scores
plt.figure(figsize=(10,6))
data[['math score', 'reading score', 'writing score']].plot(kind='hist', alpha=0.5, bins=20)
plt.title('Distribution of Math, Reading, and Writing Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.savefig(os.path.join(output_dir, 'histogram_scores.png'))
plt.show()
print("Histogram of scores saved.")
```

```python
# 2. Box Plot of Scores by Gender
plt.figure(figsize=(10,6))
sns.boxplot(x='gender', y='math score', data=data)
plt.title('Box Plot of Math Scores by Gender')
plt.savefig(os.path.join(output_dir, 'boxplot_math_gender.png'))
plt.show()
print("Boxplot of math scores by gender saved.")

plt.figure(figsize=(10,6))
sns.boxplot(x='gender', y='reading score', data=data)
plt.title('Box Plot of Reading Scores by Gender')
plt.savefig(os.path.join(output_dir, 'boxplot_reading_gender.png'))
plt.show()
print("Boxplot of reading scores by gender saved.")

plt.figure(figsize=(10,6))
sns.boxplot(x='gender', y='writing score', data=data)
plt.title('Box Plot of Writing Scores by Gender')
plt.savefig(os.path.join(output_dir, 'boxplot_writing_gender.png'))
plt.show()
print("Boxplot of writing scores by gender saved.")
```

```python
# 3. Scatter Plot of Math vs. Reading Scores
plt.figure(figsize=(10,6))
sns.scatterplot(x='math score', y='reading score', data=data, hue='gender')
plt.title('Scatter Plot of Math Score vs Reading Score')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.savefig(os.path.join(output_dir, 'scatter_math_vs_reading.png'))
plt.show()
print("Scatter plot of math vs reading scores saved.")
```

```python
# 4. Bar Plot of Average Scores by Lunch Type
plt.figure(figsize=(10,6))
lunch_avg = data.groupby('lunch')[['math score', 'reading score', 'writing score']].mean().reset_index()
sns.barplot(x='lunch', y='math score', data=lunch_avg)
plt.title('Average Math Score by Lunch Type')
plt.savefig(os.path.join(output_dir, 'barplot_avg_math_lunch.png'))
plt.show()
print("Bar plot of average math score by lunch type saved.")

plt.figure(figsize=(10,6))
sns.barplot(x='lunch', y='reading score', data=lunch_avg)
plt.title('Average Reading Score by Lunch Type')
plt.savefig(os.path.join(output_dir, 'barplot_avg_reading_lunch.png'))
plt.show()  # Display the plot inline
print("Bar plot of average reading score by lunch type saved.")

plt.figure(figsize=(10,6))
sns.barplot(x='lunch', y='writing score', data=lunch_avg)
plt.title('Average Writing Score by Lunch Type')
plt.savefig(os.path.join(output_dir, 'barplot_avg_writing_lunch.png'))
plt.show()
print("Bar plot of average writing score by lunch type saved.")
```

```python
# 5. Correlation Heatmap
plt.figure(figsize=(8,6))
corr = data[['math score', 'reading score', 'writing score']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Scores')
plt.savefig(os.path.join(output_dir, 'heatmap_correlation_scores.png'))
plt.show()
print("Correlation heatmap saved.")
```
